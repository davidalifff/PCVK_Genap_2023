```python
black = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black)
```

```python
#perhatikan koordinat titik" pt1 dan pt2
cv.rectangle(black,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt.imshow(black)
```

```python
cv.rectangle(black,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=15)
plt.imshow(black)
```

```python
cv.circle(black,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(black)
```

```python
cv.line(black,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black)
```

```python
font=cv.FONT_HERSHEY_SIMPLEX
cv.putText(black,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,0),thickness=5,lineType=cv.LINE_AA)
plt.imshow(black)
```

```python
black2=np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black2)
```

```python
vertices=np.array([[100,300],
                   [200,200],
                   [400,300],
                   [200,400]],dtype=np.int32)
vertices
```

```python
pts=vertices.reshape((-1,1,2))#fungsi reshape adalah mengubah bentuk array tanpa mengubah data
pts#dimana pada kasus ini data data akan dibentuk menjadi per kolom menambahkan kurung siku
```

```python
cv.polylines(black2,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(black2)
```
